```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
data=pd.read_csv("Book1.csv")
```

```python
%matplotlib inline
plt.scatter(data.Area,data.Price)
plt.xlabel("Area")
plt.ylabel("Price")
```

```python
reg = linear_model.LinearRegression()
reg.fit(data[['Area']],data.Price)
```

```python
reg.predict([[3300]])
```

```python
pre=pd.read_csv("Book1p.csv")
```

```python
s=reg.predict(pre)
```

```python
pre['Prices']=s
pre.to_csv("predictions.csv")
```

```python
plt.plot(data.Area,reg.predict(data[['Area']]))
```

```python
myp=pd.read_csv("predictions.csv")
```

```python
plt.scatter(myp.Area,myp.Prices)
plt.plot(myp['Area'],myp.Prices)
```
